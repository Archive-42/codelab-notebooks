```python
# take in a string
# turn the string in to a meaningful determanistic number

# take in a string
# create a sum variable
# iterate over the string
  # turn each char in to a number
  # increment sum by that number
# return sum

def my_hash(key):
  sum = 0
  for c in key:
    nc = ord(c)
    sum += nc
  return sum
storage = [None] * 10
h = my_hash("CBA")
print(h)
n = h % len(storage)
print(n)

def put(key, value):
  idx = my_hash(key) % len(storage)
  storage[idx] = value

def get(key):
  idx = my_hash(key) % len(storage)
  return storage[idx]

put("Tom", 41)




# In Python dict() is the syntax for a class. 
# d = dict()
# or d={}
dir(dict())
```
